## مقایسه عملکرد BPSK و QPSK در کانال AWGN

معیار ارزیابی عملکرد، نرخ خطای بیت (Bit Error Rate - BER) است

## مفاهیم پایه

- **BPSK**: مدولاسیون فاز دودویی که هر بیت را با یک فاز نمایش می‌دهد.
- **QPSK**: مدولاسیونی که در هر سمبل دو بیت ارسال می‌کند.
- **AWGN**: کانالی شامل نویز سفید گاوسی که مدل استاندارد نویز در مخابرات است.
- **BER**: نسبت تعداد بیت‌های اشتباه به کل بیت‌های ارسال‌شده.
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bpsk_mod(data):
    # تبدیل 0 → -1 و 1 → +1
    return 2*data - 1

def bpsk_demod(received):
    # تصمیم‌گیری بر اساس علامت سیگنال
    return (received > 0).astype(int)
```

```python
def qpsk_mod(data):
    data = data.reshape(-1, 2)
    return (2*data[:,0]-1) + 1j*(2*data[:,1]-1)

def qpsk_demod(received):
    bits = []
    for r in received:
        bits.append(1 if np.real(r) > 0 else 0)
        bits.append(1 if np.imag(r) > 0 else 0)
    return np.array(bits)
```

```python
def awgn(signal, snr_db):
    snr_linear = 10**(snr_db/10)
    signal_power = np.mean(np.abs(signal)**2)
    noise_variance = signal_power / snr_linear

    noise = np.sqrt(noise_variance/2) * (
        np.random.randn(len(signal)) + 1j*np.random.randn(len(signal))
    )
    return signal + noise
```

```python
N = 10000
snr_range = range(0, 11)

ber_bpsk = []
ber_qpsk = []

for snr in snr_range:
    data = np.random.randint(0, 2, N)

    # BPSK
    tx_bpsk = bpsk_mod(data)
    rx_bpsk = awgn(tx_bpsk, snr)
    dec_bpsk = bpsk_demod(np.real(rx_bpsk))
    ber_bpsk.append(np.mean(dec_bpsk != data))

    # QPSK
    data_qpsk = data[:N - N % 2]
    tx_qpsk = qpsk_mod(data_qpsk)
    rx_qpsk = awgn(tx_qpsk, snr)
    dec_qpsk = qpsk_demod(rx_qpsk)
    ber_qpsk.append(np.mean(dec_qpsk != data_qpsk))
```

```python
plt.semilogy(snr_range, ber_bpsk, 'o-', label='BPSK')
plt.semilogy(snr_range, ber_qpsk, 's-', label='QPSK')
plt.xlabel("SNR (dB)")
plt.ylabel("Bit Error Rate (BER)")
plt.title("BER Comparison of BPSK and QPSK over AWGN Channel")
plt.grid(True)
plt.legend()
plt.show()
```

در این بخش از نوت‌بوک، بخش‌های زیر از مقاله پیاده‌سازی شده‌اند:

1. مدولاسیون دیجیتال: BPSK و QPSK
2. کدینگ کانال ساده: Hamming (7,4)
3. شبیه‌سازی کانال نویزی AWGN

5. محاسبه نرخ خطای بیت (BER) و مقایسه نمودارها

```python
import numpy as np
import matplotlib.pyplot as plt

# مدولاسیون BPSK
def bpsk_mod(data):
    return 2*data - 1

def bpsk_demod(rx):
    return (rx > 0).astype(int)

# مدولاسیون QPSK
def qpsk_mod(data):
    data = data.reshape(-1,2)
    return (2*data[:,0]-1) + 1j*(2*data[:,1]-1)

def qpsk_demod(rx):
    bits = []
    for r in rx:
        bits.append(1 if np.real(r)>0 else 0)
        bits.append(1 if np.imag(r)>0 else 0)
    return np.array(bits)

# کانال AWGN
def awgn(signal, snr_db):
    snr = 10**(snr_db/10)
    power = np.mean(np.abs(signal)**2)
    noise_var = power/snr
    noise = np.sqrt(noise_var/2) * (
        np.random.randn(len(signal)) + 1j*np.random.randn(len(signal))
    )
    return signal + noise
```

```python
snr_range = range(0,11)
ber_bpsk = []
ber_qpsk = []

N = 10000

for snr in snr_range:
    data = np.random.randint(0,2,N)

    # BPSK
    tx = bpsk_mod(data)
    rx = awgn(tx, snr)
    dec = bpsk_demod(np.real(rx))
    ber_bpsk.append(np.mean(dec != data))

    # QPSK
    data2 = data[:N - N%2]
    tx2 = qpsk_mod(data2)
    rx2 = awgn(tx2, snr)
    dec2 = qpsk_demod(rx2)
    ber_qpsk.append(np.mean(dec2 != data2))
```

```python
plt.semilogy(snr_range, ber_bpsk, 'o-', label='BPSK')
plt.semilogy(snr_range, ber_qpsk, 's-', label='QPSK')
plt.xlabel('SNR (dB)')
plt.ylabel('BER')
plt.title('BER Comparison of BPSK and QPSK')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def hamming_encode(data_bits):
    d1, d2, d3, d4 = data_bits
    p1 = d1 ^ d2 ^ d4
    p2 = d1 ^ d3 ^ d4
    p3 = d2 ^ d3 ^ d4
    return [p1, p2, d1, p3, d2, d3, d4]

def hamming_decode(codeword):
    p1, p2, d1, p3, d2, d3, d4 = codeword
    s1 = p1 ^ d1 ^ d2 ^ d4
    s2 = p2 ^ d1 ^ d3 ^ d4
    s3 = p3 ^ d2 ^ d3 ^ d4
    syndrome = (s1 << 2) + (s2 << 1) + s3
    corrected = codeword.copy()
    if syndrome != 0:
        corrected[syndrome-1] ^= 1
    decoded = [corrected[2], corrected[4], corrected[5], corrected[6]]
    return decoded, syndrome
```

```python
data = [1,0,1,1]
code = hamming_encode(data)
# شبیه‌سازی خطای تصادفی
code[2] ^= 1  # ایجاد خطا
decoded, syndrome = hamming_decode(code)

print("داده اصلی: ", data)
print("کد Hamming: ", code)
print("سندروم خطا: ", syndrome)
print("داده تصحیح‌شده: ", decoded)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# تعداد بیت‌ها
N = 10000

# دامنه SNR
snr_range = range(0, 11)

# مدولاسیون BPSK
def bpsk_mod(data):
    return 2*data - 1

def bpsk_demod(rx):
    return (rx > 0).astype(int)

# کانال AWGN
def awgn(signal, snr_db):
    snr = 10**(snr_db/10)
    power = np.mean(np.abs(signal)**2)
    noise_var = power/snr
    noise = np.sqrt(noise_var/2) * np.random.randn(len(signal))
    return signal + noise

# Hamming(7,4) - کدگذاری
def hamming_encode(data_bits):
    d1, d2, d3, d4 = data_bits
    p1 = d1 ^ d2 ^ d4
    p2 = d1 ^ d3 ^ d4
    p3 = d2 ^ d3 ^ d4
    return np.array([p1, p2, d1, p3, d2, d3, d4])

# Hamming(7,4) - تصحیح خطا
def hamming_decode(codeword):
    p1, p2, d1, p3, d2, d3, d4 = codeword
    s1 = p1 ^ d1 ^ d2 ^ d4
    s2 = p2 ^ d1 ^ d3 ^ d4
    s3 = p3 ^ d2 ^ d3 ^ d4
    error_pos = s3*4 + s2*2 + s1*1
    decoded = np.array([d1, d2, d3, d4])
    if error_pos != 0:
        codeword[error_pos-1] ^= 1  # اصلاح بیت اشتباه
        # دوباره بیت‌ها را بازیابی کن
        p1, p2, d1, p3, d2, d3, d4 = codeword
        decoded = np.array([d1, d2, d3, d4])
    return decoded

# محاسبه BER بدون و با Hamming
ber_no_hamming = []
ber_hamming = []

for snr in snr_range:
    errors_no = 0
    errors_ham = 0
    for _ in range(N//4):  # 4 بیت در هر بلوک
        data = np.random.randint(0,2,4)
        # BPSK بدون Hamming
        tx = bpsk_mod(data)
        rx = awgn(tx, snr)
        dec = bpsk_demod(rx)
        errors_no += np.sum(dec != data)

        # BPSK با Hamming
        code = hamming_encode(data)
        tx_code = bpsk_mod(code)
        rx_code = awgn(tx_code, snr)
        dec_code = bpsk_demod(rx_code)
        dec_data = hamming_decode(dec_code)
        errors_ham += np.sum(dec_data != data)

    ber_no_hamming.append(errors_no/N)
    ber_hamming.append(errors_ham/N)

# رسم نمودار
plt.figure(figsize=(8,5))
plt.semilogy(list(snr_range), ber_no_hamming, 'o-', label='بدون Hamming')
plt.semilogy(list(snr_range), ber_hamming, 's-', label='با Hamming')
plt.xlabel('SNR (dB)')
plt.ylabel('BER')
plt.title('تاثیر Hamming Code بر نرخ خطای بیت (BER)')
plt.grid(True)
plt.legend()
plt.show()
```

## بخش Equalizer و مقایسه BER

در این بخش، یک **Equalizer ساده LMS** برای سیگنال BPSK پیاده‌سازی شده است.  
هدف این است که نشان دهیم **چطور Equalizer می‌تواند اثر نویز و اختلالات کانال را کاهش دهد** و نرخ خطای بیت (BER) را بهبود ببخشد.  

سپس، **BER سیگنال BPSK با و بدون Equalizer** در SNRهای مختلف محاسبه و نمودار مقایسه رسم می‌شود.  
این شبیه‌سازی اهمیت روش‌های Equalization را در شبکه‌های مخابراتی نشان می‌دهد.
</div>

```python
# LMS Equalizer ساده برای سیگنال BPSK
def lms_equalizer(rx, mu=0.01, taps=5):
    """
    rx: سیگنال دریافتی
    mu: نرخ یادگیری
    taps: تعداد تاپ‌های فیلتر
    """
    N = len(rx)
    w = np.zeros(taps)
    y = np.zeros(N)
    for n in range(taps, N):
        x = rx[n-taps:n][::-1]
        y[n] = np.dot(w, x)
        e = np.sign(rx[n]) - y[n]  # خطای ساده
        w += mu * e * x
    return y
```

```python
import numpy as np

# تعداد نمونه‌ها
N = 10000

# دامنه SNR
snr_range = range(0, 11)

# مدولاسیون BPSK
def bpsk_mod(data):
    return 2*data - 1

def bpsk_demod(rx):
    return (rx > 0).astype(int)

# کانال AWGN
def awgn(signal, snr_db):
    snr = 10**(snr_db/10)
    power = np.mean(np.abs(signal)**2)
    noise_var = power/snr
    noise = np.sqrt(noise_var/2) * (
        np.random.randn(len(signal)) + 1j*np.random.randn(len(signal))
    )
    return signal + noise

# Equalizer ساده LMS (ورودی complex128)
def lms_equalizer(rx, mu=0.01, taps=5):
    N = len(rx)
    w = np.zeros(taps, dtype=complex)
    y = np.zeros(N, dtype=complex)
    for n in range(taps, N):
        x = rx[n-taps:n][::-1]
        y[n] = np.dot(w, x)
        e = np.sign(np.real(rx[n])) - np.real(y[n])  # فقط بخش حقیقی
        w += mu * e * x
    return y

# محاسبه BER
ber_bpsk = []
ber_bpsk_eq = []

for snr in snr_range:
    data = np.random.randint(0, 2, N)
    tx = bpsk_mod(data)
    rx = awgn(tx, snr)

    # بدون Equalizer
    dec = bpsk_demod(np.real(rx))
    ber_bpsk.append(np.mean(dec != data))

    # با Equalizer
    rx_eq = lms_equalizer(rx)
    dec_eq = bpsk_demod(np.sign(np.real(rx_eq)))
    ber_bpsk_eq.append(np.mean(dec_eq != data))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.semilogy(list(snr_range), ber_bpsk, 'o-', label='BPSK (بدون Equalizer)')
plt.semilogy(list(snr_range), ber_bpsk_eq, 's-', label='BPSK (با Equalizer)')
plt.xlabel('SNR (dB)')
plt.ylabel('BER')
plt.title('BER Comparison of BPSK With/Without Equalizer')
plt.grid(True)
plt.legend()
plt.show()
```

### تحلیل نتایج Equalizer

1. استفاده از LMS Equalizer باعث کاهش نرخ خطای بیت می‌شود، به خصوص در SNR پایین.
2. نمودار نشان می‌دهد که Equalizer می‌تواند اثر نویز و اختلالات کانال را تا حدی جبران کند.
3. این بخش شبیه‌سازی اهمیت روش‌های Equalization را در شبکه‌های مخابراتی نشان می‌دهد.
4. ترکیب مدولاسیون، کدینگ و Equalization بهبود قابل توجهی در کیفیت دریافت داده ایجاد می‌کند.
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# تعداد بیت‌ها
N = 20000
snr_range = range(0, 11)

# BPSK مدولاسیون
def bpsk_mod(data):
    return 2*data - 1

def bpsk_demod(rx):
    return (rx > 0).astype(int)

# کانال با یک مسیر اصلی و نویز AWGN
def awgn_channel(tx, snr_db):
    snr = 10**(snr_db/10)
    power = np.mean(tx**2)
    noise_var = power/snr
    noise = np.sqrt(noise_var) * np.random.randn(len(tx))
    return tx + noise

# Equalizer LMS
def lms_equalizer(rx, tx, mu=0.01, taps=5):
    N = len(rx)
    w = np.zeros(taps)
    y = np.zeros(N)
    rx_padded = np.concatenate([np.zeros(taps), rx])
    for n in range(taps, N+taps):
        x = rx_padded[n-taps:n][::-1]
        y[n-taps] = np.dot(w, x)
        e = tx[n-taps] - y[n-taps]
        w += 2*mu*e*x
    return y

# محاسبه BER
ber_no_eq = []
ber_eq = []

for snr in snr_range:
    errors_no = 0
    errors_eq = 0
    for _ in range(N//1000):
        data = np.random.randint(0,2,1000)
        tx = bpsk_mod(data)
        # کانال + نویز
        rx = awgn_channel(tx, snr)

        # بدون Equalizer
        dec = bpsk_demod(rx)
        errors_no += np.sum(dec != data)

        # با Equalizer
        rx_eq = lms_equalizer(rx, tx)
        dec_eq = bpsk_demod(rx_eq)
        errors_eq += np.sum(dec_eq != data)

    ber_no_eq.append(errors_no/N)
    ber_eq.append(errors_eq/N)

# رسم نمودار BER
plt.figure(figsize=(8,5))
plt.semilogy(list(snr_range), ber_no_eq, 'o-', label='BPSK بدون Equalizer')
plt.semilogy(list(snr_range), ber_eq, 's-', label='BPSK با Equalizer')
plt.xlabel('SNR (dB)')
plt.ylabel('BER')
plt.title('تاثیر Equalizer در کانال AWGN')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# تعداد بیت‌ها
N = 50000
snr_range = range(0, 11)

# BPSK مدولاسیون
def bpsk_mod(data):
    return 2*data - 1

def bpsk_demod(rx):
    return (rx > 0).astype(int)

# کانال چندمسیره با نویز AWGN
def multipath_channel(tx, taps=[1.0, 0.5, 0.3], delays=[0, 1, 2]):
    rx = np.zeros(len(tx) + max(delays))
    for tap, delay in zip(taps, delays):
        rx[delay:delay+len(tx)] += tap * tx
    return rx[:len(tx)]

def awgn(signal, snr_db):
    snr = 10**(snr_db/10)
    power = np.mean(np.abs(signal)**2)
    noise_var = power/snr
    noise = np.sqrt(noise_var) * np.random.randn(len(signal))
    return signal + noise

# Equalizer LMS واقعی
def lms_equalizer(rx, tx, mu=0.01, taps=7):
    N = len(rx)
    w = np.zeros(taps)
    y = np.zeros(N)
    rx_padded = np.concatenate([np.zeros(taps), rx])
    for n in range(taps, N+taps):
        x = rx_padded[n-taps:n][::-1]
        y[n-taps] = np.dot(w, x)
        e = tx[n-taps] - y[n-taps]
        w += 2*mu*e*x
    return y

# محاسبه BER
ber_no_eq = []
ber_eq = []

for snr in snr_range:
    errors_no = 0
    errors_eq = 0
    for _ in range(N//5000):
        data = np.random.randint(0,2,5000)
        tx = bpsk_mod(data)
        # کانال چندمسیره + نویز
        rx = multipath_channel(tx)
        rx = awgn(rx, snr)

        # بدون Equalizer
        dec = bpsk_demod(rx)
        errors_no += np.sum(dec != data)

        # با Equalizer
        rx_eq = lms_equalizer(rx, tx)
        dec_eq = bpsk_demod(rx_eq)
        errors_eq += np.sum(dec_eq != data)

    ber_no_eq.append(errors_no/N)
    ber_eq.append(errors_eq/N)

# رسم نمودار BER
plt.figure(figsize=(8,5))
plt.semilogy(list(snr_range), ber_no_eq, 'o-', label='BPSK بدون Equalizer')
plt.semilogy(list(snr_range), ber_eq, 's-', label='BPSK با Equalizer')
plt.xlabel('SNR (dB)')
plt.ylabel('BER')
plt.title('تاثیر Equalizer در کانال چندمسیره')
plt.grid(True)
plt.legend()
plt.show()
```
